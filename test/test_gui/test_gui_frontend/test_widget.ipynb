```python
import ipyvuetify as v
import ipywidgets as widgets
import pandas as pd
import numpy as np
EXTENSION = '.csv'


data = pd.read_csv('075773_treated_xye' + EXTENSION, sep=',')

data = data.apply(pd.to_numeric, errors='coerce')
data = data.dropna()

q = np.array(data.iloc[:, 0])
I = np.array(data.iloc[:, 1])
dI = np.array(data.iloc[:, 2])

from plotly.graph_objects import FigureWidget, Scattergl

fig = FigureWidget(data=Scattergl(
    x=q,
    y=I,
    mode='markers',
    marker=dict(size=14, color='rgba(219, 64, 82, 0.8)')
))

fig.update_layout(
    title='2D Scatter Plot',
    xaxis_title='X Axis Title',
    yaxis_title='Y Axis Title',
    width=300, 
    height=300,
    autosize=True,
    margin={
        't': 0,
        'b': 0,
        'l': 0,
        'r': 0
    },
)

def create_text_fields():
    return [
        v.TextField(label='starts', value='{b → bstart}'),
        v.TextField(label='lims', value='{0.7 minq < x0 < 1.5 maxq, 0.5 < b/bstart < 5}'),
        v.Html(tag='div', children=['qNEW(q) = (1 - (3 λ^2)q^2 / 32π^2)q/λ → 1.54']),
    ]

def create_buttons():
    return [
        v.Btn(children=['Watch messages']),
        v.Btn(children=['Watch process']),
    ]

layout = v.Container(
    style_="width: 1000px; height: 1500px;",
    
    children=[
        v.Row(
            class_="d-flex",
            children=[
                v.Col(
                    style_="width: 100%",
                    class_="d-flex flex-column justify-center",
                    children=[
                        v.Tabs(
                            v_model=0,
                            children=[
                                         v.Tab(children=["Text Fields"]),
                                         v.Tab(children=["Buttons"]),
                                     ] + [
                                         v.TabItem(
                                             class_="mt-2",
                                             children=[v.Card(children=create_text_fields())]
                                         ),
                                         v.TabItem(
                                             children=[v.Card(children=create_buttons())]
                                         ),
                                     ]
                        )
                    ]
                )
            ]
        ),
        
        v.Row(
            children=[
                v.Col(
                    children=[v.Select(label='Set', items=['1st set', '2nd set'], v_model='1st set')]
                )
            ]
        ),
        v.Divider(),
        v.Row(
            children=[
                v.Col(
                    style_="width: 50%",
                    children=[v.Card(
                        children=[
                            v.CardTitle(children=['Text Fields']),
                            v.CardText(children=create_text_fields())
                        ]
                    )],
              ), 
            v.Col(
                style_="width: 50%",
                children=[v.Card(
                    children=[
                        v.CardTitle(children=['Buttons']),
                        v.CardText(children=create_buttons())
                    ]
                )],
          ),
            
            ]
        ),
        v.Divider(),
        v.Row(
            children=[
                v.CardTitle(children=['Additional Controls']),
                v.CardText(
                    children=[
                        v.Slider(label='b0', min=0, max=0.1, step=0.001, v_model=0.006),
                        v.Checkbox(label='Search for Sponge & FI phases'),
                        v.Btn(children=['Start'], color='primary', large=True)
                    ]
                )
            ]
        ),
        v.Row(
            children=[
            v.Col(
                style_="width: 50%",
                children=[
                        v.CardTitle(children=['Buttons']),
                        fig]
          ),
            v.Col(
                style_="width: 50%",
                children=[
                        v.CardTitle(children=['Buttons']),
                        fig]
          ),
            
            ]
        ),
    ]
)

# Display the layout
display(layout)
```
